```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

import re
import seaborn as sns

pd.set_option('display.max_columns', None)
keys = {
    "Anxiety": ["anxiety", "anxious", "overwhelm", "stress"],
    "Depression": ["depression", "depressed"],
    "Autism":["autism"], "bipolar": ["bipolar"],
    "Suicide":["suicide", "suicidal"],
    "PanicAttack": ["panic"],
    "OCD": ["overthink", "ocd"],
    "ADHD": ["adhd"],
    }

def find_keywords_in_data(sentence, key):
    for value in keys[key]:
        if value in str(sentence).lower():
            return 1
    
    return 0


df = pd.read_excel("ADHD.xlsx")

removed_columns = ["specify", "home_language", "nbt_completed", "nbt_year", "nbt_did_math"]
removed_columns += [f"bdi1_item_{i}" for i in range(1,22)]
removed_columns += [f"audit1_item_{i}" for i in range(1,10)]
removed_columns += [f"aas1_item_{i}" for i in range(1,10)]
removed_columns += [f"asrs1_item_{i}" for i in range(1,19)]
removed_columns += [f"bai1_item_{i}" for i in range(1,22)]

df.drop(columns=removed_columns, inplace=True)
df.reset_index()
renamed_columns = {
    'have_you_ever_experienced_any_mental_health_difficulties_or_symptoms_before_starting_university_e_g_in_primary_or_high_school': 'PreUniMentalHealthIssues',
    'if_yes_please_list_these_difficulties_and_or_symptoms': 'ListedDifficulties',
    'if_you_have_ever_experienced_difficulties_and_or_symptoms_of_a_mental_illness_how_old_were_you_when_this_started': 'MentalIllnessStartAge',
    'have_you_ever_been_diagnosed_with_a_mental_illness': 'MentalIllnessDiagnosis',
    'if_you_have_been_diagnosed_formally_or_informally_please_list_the_diagnosis_diagnoses': 'ListedDiagnoses',
    'was_this_diagnosis_made_before_or_after_you_left_high_school': 'DiagnosisTiming',
    'if_you_have_been_diagnosed_with_a_mental_illness_at_what_age_was_this': 'DiagnosisAge',
    'have_you_ever_used_prescribed_psychiatric_medication_for_a_mental_illness_or_symptoms_of_one': 'UsedPsychMedication',
    'are_you_currently_using_prescribed_psychiatric_medication_for_a_mental_illness_or_symptoms_of_one': 'CurrentPsychMedicationUse',
    'have_you_ever_been_to_therapy_or_counselling_for_a_mental_illness_or_symptoms_of_one': 'TherapyExperience',
    'are_you_currently_in_therapy_or_counselling_for_a_mental_illness_or_symptoms_of_one': 'CurrentTherapyStatus'
    }
df.rename(columns=renamed_columns, inplace=True)
max_age = df["age"].max()
df['MentalIllnessStartAge'] = df['MentalIllnessStartAge'].str.extract('(\d+)', expand=False)
df['MentalIllnessStartAge'] = df['MentalIllnessStartAge'].apply(lambda x: np.nan if (pd.isnull(x) or not pd.isnull(x) and int(x) > max_age) else int(x))
df['DiagnosisAge'] = df['DiagnosisAge'].str.extract('(\d+)', expand=False)
df['DiagnosisAge'] = df['DiagnosisAge'].apply(lambda x: np.nan if (pd.isnull(x) or not pd.isnull(x) and int(x) > max_age) else int(x))

for ill in keys.keys():
    df[f'PreUni{ill}'] = df["ListedDifficulties"].apply(find_keywords_in_data, args=(ill,))
    df[ill] = df["ListedDiagnoses"].apply(find_keywords_in_data, args=(ill,))

    # # TODO: reset index fixing.

new_df = df.drop(columns=["ListedDifficulties", "ListedDiagnoses"])
new_df["MentalIllnessDiagnosis"] = new_df["MentalIllnessDiagnosis"].replace(
    {
        "not formally diagnosed": "not formally",
        "yes, formally diagnosed by a doctor or mental health professional": "yes formally"
    }
)

features = ["UsedPsychMedication", "CurrentPsychMedicationUse", "TherapyExperience", "CurrentTherapyStatus", "PreUniMentalHealthIssues", "MentalIllnessDiagnosis", "sex"]

ohe = OneHotEncoder(sparse_output=False)

for feature in features:
    transformed_data = ohe.fit_transform(new_df[[feature]])
    transformed_df = pd.DataFrame(transformed_data, columns=ohe.get_feature_names_out([feature]))
    
    new_df = new_df.drop([feature], axis=1)
    new_df = pd.concat([new_df, transformed_df], axis=1)


label_encoder = LabelEncoder() 
new_df["DiagnosisTiming"] = label_encoder.fit_transform(new_df["DiagnosisTiming"])
new_df.drop(columns=["CurrentPsychMedicationUse_not applicable","UsedPsychMedication_not applicable","TherapyExperience_not applicable","CurrentTherapyStatus_not applicable", "DiagnosisAge", "MentalIllnessStartAge", "aas_change"], inplace=True)


new_df
```

```python
corr_matrix = new_df.corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool_))

to_drop = [column for column in upper.columns if any(upper[column] > 0.90)]

new_df = new_df.drop(new_df[to_drop], axis=1)
train_df, test_df = train_test_split(new_df, test_size=0.3)
pca = PCA(n_components=2)
df_pca = pca.fit_transform(new_df)

kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit_transform(df_pca)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca["Cluster"] = kmeans.labels_
score = silhouette_score(df_pca, kmeans.labels_)
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue="Cluster").set(title=f'Score:{score}')
```

```python


from lazypredict import LazyClassifier

X = new_df.drop(["PreUniADHD", "ADHD"], axis=1)
y = new_df['ADHD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

models
```

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Decision Tree classifier F1 score:", f1_score(y_test, y_pred, average="weighted"))
```

```python
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier()
scores = cross_val_score(clf, X, y, cv=5, scoring='f1_weighted')

print("F1 Score with Cross Validation:", scores.mean())
```
